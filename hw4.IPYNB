- id - Уникальный ID для каждого дома
- date - Дата продажи дома
- price - Стоимость продажи дома
- bedrooms - Кол-во спален
- bathrooms - Кол-во ванных комнат (0.5 - туалет без душа)
- sqft_living - Кв. метры жилые
- sqft_lot - Кв. метры общие
- floors - Кол-во этажей
- waterfront - Есть набержная или нет
- view - Значение от 0 до 4 насколько хороший вид
- condition - Значение от 1 до 5 насколько хорошее состояние
- grade - Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна
- sqft_above - Кв. метры дома, которые находятся выше земли
- sqft_basement - Кв. метры дома, которые находятся ниже земли
- yr_built - Год постройки дома
- yr_renovated - Год ремонта дома
- zipcode - Индекс
- lat - Широтма
- long - Долгота
- sqft_living15 - Кв. метры жилой площади у 15 соседей
- sqft_lot15 - Кв. метры общей площади у 15 соседей

```python
df = df.drop(columns=['sqft_basement','sqft_above',	'zipcode','lat','long','sqft_living15','sqft_lot15'])
```

```python
import numpy as np
```

```python
pricehous = df.corr(numeric_only=True)
pricehous = np.round(pricehous,2)
pricehous = np.abs(pricehous)
pricehous
```

```python
plt.figure()
sns.heatmap(pricehous, annot = True, linewidths=0.5 , cmap='coolwarm')
```

Наибольшее влияние на цену оказывают жилая площадь и количество ванных комнат.

```python
# Создание переменных для каждой категории
price_data = df.groupby(['price']).size()
bedrooms_data = df.groupby(['bedrooms']).size()
SQL_data = df.groupby(['sqft_living']).size()

# Создание многорядной столбчатой диаграммы
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

axs[0].bar(price_data.values, price_data.index)
axs[0].set_title('Распределение по ценовой категории')

axs[1].bar(bedrooms_data.index, bedrooms_data.values)
axs[1].set_title('Распределение по количеству спален')

axs[2].bar(SQL_data.values, SQL_data.index)
axs[2].set_title('Распределение по жилой площади')

plt.show()
```
